```python
import requests 
import pandas as pd
```

```python
url = "https://beta.adalab.es/resources/apis/pelis/pelis.json"
llamada = requests.get (url)
```

```python
llamada.status_code
```

```python
datos = llamada.json ()
```

```python
datos
```

```python
datos_insercion_peliculas = []

for pelicula in datos:
    if 'subtitulos' in pelicula:
        del pelicula['subtitulos']

for pelicula in datos:
    datos_insercion_peliculas.append(tuple(pelicula.values()))
print(datos_insercion_peliculas)

```

```python
display(datos_insercion_peliculas)
```

```python
for pelicula in datos:
    tupla = (
        pelicula['id'],
        pelicula['titulo'],
        pelicula['año'],
        pelicula['duracion'],
        pelicula['genero'],
        pelicula['adultos'] 
    )
    datos_insercion_peliculas.append(tupla)

print(datos_insercion_peliculas)
```
